# Transform Data Using function and Mapping

```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
 'Pastrami', 'corned beef', 'Bacon',
 'pastrami', 'honey ham', 'nova lox'],
 'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {
'bacon': 'pig',
'pulled pork': 'pig',
'pastrami': 'cow',
'corned beef': 'cow',
'honey ham': 'pig',
'nova lox': 'salmon'
}
lowerdcase=data['food'].str.lower()
lowerdcase
```

```python
data['animal']=lowerdcase.map(meat_to_animal)
data
```

```python
data['food'].map(lambda x:meat_to_animal[x.lower()])
```

# Discritization and Bining

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins=[18,25,35,60,100]
group=pd.cut(ages,bins)
group
```

```python
group.codes
```

```python
group.categories
```

```python
pd.value_counts(group)
```

```python
pd.cut(ages,[18,25,35,60,100],right=True)
```

```python
group_names=['youth','young age','middleage','senior']
pd.cut(ages,bins,labels=group_names)
```
